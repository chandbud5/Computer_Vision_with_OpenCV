# OpenCV

### Importing OpenCV

```
import cv2
```

**Matplotlib read image as RGB and OpenCV read image channels as BGR**

Reading image with OpenCV

```
img = cv2.imread("/content/232876.jpg")
type(img)
```

```
plt.imshow(img)
```

```
img_fix = cv2.cvtColor(img,
                       cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

**Reading image in grayscale**

```
img_gray = cv2.imread("/content/232876.jpg", cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
plt.imshow(img_gray)
```

**Resizing image**

```
img_size = cv2.resize(img_fix, (600, 800))
plt.imshow(img_size)
```

**Resizing with fix ratio**

```
width_r = 0.5
height_r = 0.5
img_size = cv2.resize(img_fix, (0,0), img_fix, width_r, height_r)
```

```
plt.imshow(img_size)
```

**Flip on Horizontal Axis**

```
img2 = cv2.flip(img_fix, 0) # 0 for horizontal axis
plt.imshow(img2)
```

```
img2 = cv2.flip(img_fix, 1) # 1 for vertical axis
plt.imshow(img2)
```

Flipping across both axis

```
img2 = cv2.flip(img_fix, -1) # -1 for horizontal and vertical both axis
plt.imshow(img2)
```

```
img2 = plt.figure(figsize=(10,8))
ip = img2.add_subplot(111)
ip.imshow(img_fix)
```
